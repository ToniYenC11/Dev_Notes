# Cache Memory

- Reduces average time needed to retrieve data from main memory.

- Effectiveness lies on the principle of *locality of reference*.
    
    - **Locality of reference** - Tendency of a computer program to access instructions whose addresses are near each other. This is shown by *loops* and *subroutines*.

    - In case of loops in program control processing unit repeatedly refers to the set of instructions that constitute the loop.
    - In case of subroutine calls, every time the set of instructions are fetched from memory.
    - References to data items also get localized that means same data item is referenced again and again.

    

- Cache memory stores data close to the CPU, which helps speed up processing. It's much faster than the main memory (RAM). 

### Principles of Cache

- It is fast, but can only hold small amount of data.

- The CPU checks the cache first due to how quick it would be. A data found in the cache is a **cache hit**.
- If not in the cache, it looks in the RAM where if it is found, it moves the data to the cache. This is a **cache miss**.

### Types of Cache Memory

- **L1** - Present inside the processor. Size is from 2KB to 64 KB.

- **L2** - May be present inside or outside the CPU. If not inside, it can be shared between cores depending on the architecture. Size is 256KB to 512 KB.

- **L3** - Present outside the CPU and shared by all the cores of the CPU. Increases performance of L2 and L1 cache. Size is 1MB - 8 MB.

## Main Memory

### Random Access Memory

- Stores data temporarily while computer is running. Processes information faster.

- **Volatile** - The memory is lost when the system is turned off.
