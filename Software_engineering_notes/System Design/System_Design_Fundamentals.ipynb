# Key Memory and Storage Systems

## Cache Memory

- Reduces average time needed to retrieve data from main memory.

- Effectiveness lies on the principle of *locality of reference*.
    
    - **Locality of reference** - Tendency of a computer program to access instructions whose addresses are near each other. This is shown by *loops* and *subroutines*.

    - In case of loops in program control processing unit repeatedly refers to the set of instructions that constitute the loop.
    - In case of subroutine calls, every time the set of instructions are fetched from memory.
    - References to data items also get localized that means same data item is referenced again and again.

    

- Cache memory stores data close to the CPU, which helps speed up processing. It's much faster than the main memory (RAM). 

### Principles of Cache

- It is fast, but can only hold small amount of data.

- The CPU checks the cache first due to how quick it would be. A data found in the cache is a **cache hit**.
- If not in the cache, it looks in the RAM where if it is found, it moves the data to the cache. This is a **cache miss**.

### Types of Cache Memory

- **L1** - Present inside the processor. Size is from 2KB to 64 KB.

- **L2** - May be present inside or outside the CPU. If not inside, it can be shared between cores depending on the architecture. Size is 256KB to 512 KB.

- **L3** - Present outside the CPU and shared by all the cores of the CPU. Increases performance of L2 and L1 cache. Size is 1MB - 8 MB.

## Main Memory

### Random Access Memory

- Stores data temporarily while computer is running. Processes information faster.

- **Volatile** - The memory is lost when the system is turned off.

#### **Static RAM**

- Uses CMOS to hold memory, with 4-6 transistors and uses clocks.

- More expensive and requires more power at higher frequencies, but faster.
    - It cannot store more bits per chip. For the same amount of memory store in DRAM, SRAM would require one more chip.

#### **DRAM**

- Used for the main memory. Uses one transistor and one capacitor.

- Slower, but can store more bits per chip and requires less power, hence, cheaper.
    - It has more power consumption vs DRAM, but at higher fequencies, the consumption of both RAMs are comparable.

##### Synchronous DRAM (SDRAM)

-  Any DRAM where the operation of its external pin interface is coordinated by an externally supplied clock signal.

##### Graphics DDRAM (GDDR DRAM/SRAM)

- Synchronous DRAM for applications with high bandwidth.

- Uses parallel processing.

#### Double Data Rate

- The SRAM and DRAM have a variation that has double the speed.

### Read-Only Memory

- Permanently store data that does not need to be modified.

- **Non-volatile** - retains stored infromation when power is off.
    - Typically used for the BIOS (contains instructions for booting the computer) as well as the firmware of other hardware devices.

#### BIOS

- Computer program first to run when a computer is booted.

- It has the following features:

    - **POST**: Power on Self-Test is the process performed by BIOS immediately after the computer system starts to check any hardware-related issues. if any error is found in this process, the system alerts through flashing LEDs, audible beeps, or error messages on the monitor, all of which are technically referred to as POST codes, beep codes, and on-screen POST error messages, respectively. 

    - **Bootstrap Loader**: The bootstrap program stored in ROM is loaded in the hard drive, as soon as the computer starts. The bootstrap loader locates the operating system and passes the control to it. 

    - **Bootstrap Driver**: Bootstrap Driver configures basic hardware so that the system can run. They are low-level drivers and give basic information about computer hardware.

    - **CMOS Setup**: Complementary Metal-Oxide Semiconductor (CMOS) chip, saves some system information, such as the time, date, and other essential system settings. CMOS chip is powered by the battery located in the motherboard. It also stores some information about hardware. The chip's content is copied to the RAM(main memory) of the system to increase the speed since the CMOS chip is quite slow. It is because the BIOS and CMOS, setup configure the systems crucial information like the password, date, and time.

    - **Boot device selection**: One of the CMOS's most important roles is, it can alter the device boot process. This is important for system restoration because the CMOS may need to change boot priority from the hard drive to the optical drive or flash drive to launch the operating system installer or adjust which hard drive to load the operating system from.

- Nowadays, it is stored in the **flash memory** so that it can be rewritten wtihout removing the chip from the motherboard.
    - Also to easily update the BIOS firmware

- BIOS evolved into the **Unified Extensible Firmware Interface** (UEFI).  

#### Firmware

- Any software that provides low-level control of computing device hardware.

- For a more complex device, firmware may provide relatively low-level control as well as hardware abstraction services to higher-level software such as an operating system.

#### Types of ROM

##### **Programmable Read-only (PROM)**

- Can be programmed. Once programmed, the data and instructions cannot be changed.

##### **Erasable PROM (EPROM)**

- Can be repogrammed. 

- To delete data, expose to UV light.

##### **Electrically EPROM (EEPROM)**

- Data can be erased by applying an electric field.

- Will only erase portions of the chip.

##### **Mask ROM (MROM)**

- Mask ROM is a kind of read-only memory, that is masked off at the time of production
