## More Distributions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#~ Normal Distribution - Symmetrical in the center, and the probability is never zero
#~ Under a normal distribution, 68% falls within 1 standard deviation. 95% falls within 2 standard deviations.
#& Standard Normal Distribution - Mean=0, Standard Deviation=1
#* Involves the p-value calculations.

from scipy.stats import norm
print(norm.cdf(154,161,7)) #* Number of interest 154. Mean = 161 Standard Dev = 7
print(1-norm.cdf(154,161,7))
print(norm.cdf(157,161,7)-norm.cdf(154,161,7))
print(norm.ppf(0.9,161,7)) #* What height are 90% of women shorter than? Mean 161, std 7.

#* random numbers
print(norm.rvs(161,7,size=10))


```

```python

#~ Central Limit Theorem - A sampling distribution becomes closer to a normal distribution as the number of trials increase
sided_die_36 = pd.Series([i for i in range(1,37)])
sample_list=[]
for i in range (1,1001):
    a = sided_die_36.sample(5,replace = True).mean() #* Roll die 5 times, then take the mean. Doe for 1000 trials.
    sample_list.append(a)
plt.hist(sample_list,bins=30)
```

```python
from scipy.stats import poisson
from scipy.stats import expon
#~ Poission process - Events appear to happen at a certain rate but completely at random
#& Poisson Distribution
print(poisson.pmf(5,8)) #* more than 5 adoptions per week given an average 8 adoptions per week
poisson.rvs(8,size=10) #* CLT still applies

#& Exponential distribution - probability of time between Poisson events
print(expon.cdf(1,scale=2)) #* lambda = 1/2 = 0.5

#& t-distribution - Similar to a normal distribution except observations fall further from the mean

#& Log-normal distribution - Variable whose logartithm is normally distributed
```
