## Quantitative Comparions

```python
import pandas as pd
import matplotlib.pyplot as plt
olympics=pd.read_csv('/home/toni/Documents/Programming_Projects/Data Science Projects/Datasets/Tokyo_2021_dataset.csv')
display(olympics.head())

#& Bar charts
olympic_asia=olympics[olympics['Continent']=='Asia']
fig, ax = plt.subplots()
ax.bar(olympic_asia['Team/NOC'],olympic_asia['Gold Medal'],label='Gold')
ax.set_xticklabels(olympic_asia['Team/NOC'], rotation=90) #* rotate to properly display countries\
ax.set_ylabel('Number of Medals')
ax.bar(olympic_asia['Team/NOC'],olympic_asia['Silver Medal'], bottom = olympic_asia['Gold Medal'],label='Silver')
ax.bar(olympic_asia['Team/NOC'],olympic_asia['Bronze Medal'], bottom = olympic_asia['Gold Medal']+olympic_asia['Silver Medal'],label='Bronze')
ax.legend() #* Shows label color
plt.style.available                 
```

```python
participants=pd.read_excel('/home/toni/Documents/Programming_Projects/Data Science Projects/Datasets/Athletes.xlsx')
display(participants.head())

#~ Histogram
fig, ax= plt.subplots()
ax.hist(participants[participants['Discipline']=='Basketball']['Discipline'])
ax.hist(participants[participants['Discipline']=='Rowing']['Discipline'])
ax.set_ylabel('Counts')
ax.set_xlabel('Sports')

#~ Adding error bars
ax.bar("Gold",olympic_asia['Gold Medal'].mean(), yerr=olympic_asia['Gold Medal'].std())
ax.bar("Silver",olympic_asia['Silver Medal'].mean(), yerr = olympic_asia['Silver Medal'].std())    

#&errorbars can also be used in lihnecharts using ax.errorbar()
```

# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

bmi=pd.read_csv('/home/toni/Documents/Programming_Projects/Data Science Projects/Datasets/bmi.csv')
display(bmi.head())
sns.scatterplot(x=bmi['Height'],y=bmi['Weight'])
plt.show()

#~ count plots
sns.countplot(x=bmi['Gender'])
plt.show()
print(bmi.groupby('Gender').count())
```

```python

#~ relplot - relational plot to compare variables using a third variable
weight = bmi['Weight']
height = bmi['Height']
sns.relplot(x=weight,y=height, kind='scatter',col=bmi['Gender']) #* replace col with row if vertical
```
