# Matplotlib

### Basic Plots

```python
import matplotlib.pyplot as plt
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
import data

# * line plot - good for continuous variable vs time (months, year, etc.)
plt.plot(year,pop)
plt.title("A random graph")
plt.xlabel("Year")
plt.ylabel("Population in million")
plt.show()

# *Scatter - Good for two continuous variables with each other
plt.scatter(year,pop)
plt.title("A random graph but in scatter")
plt.xlabel("Year")
plt.ylabel("Population in million")
plt.show()

# &xscale - Set x axis to logarithmic
plt.xscale('log')
plt.scatter(data.gdp_cap,data.life_exp)
plt.show()
```

### Histogram 

Gets idea about the distribution, and explores the dataset.\


```python
plt.hist(data.gdp_cap) # * default bins is 10
plt.title("Histogram of gdp_cap")
plt.xlabel('gdp_cap')
```

### Customization

Depends on data and visual presentation of the data

```python
plt.xscale('log')
plt.scatter(data.gdp_cap,data.life_exp) 
# & axis labels using xlabel and ylabel\
plt.ylabel('count')
plt.xlabel('gdp_cap')
# & Add plot title
plt.title("Histogram of gdp_cap")
# & Scale the axis with ticks (yticks and xticks)
plt.yticks([x*10 for x in range (0,10)],['{0} count'.format(x*10) for x in range(0,10) ]) 
# Second argument of yticks is for the label of the y-axis

# & color can be used to color various data points. Use c 
# & alpha changes the opacity of the points
# & Draw grid lines using plt.grid(True)
col='purple'
plt.scatter(data.gdp_cap,data.life_exp,c=col, alpha=0.5) 
plt.grid(True)
```

### Pandas

```python
import pandas as pd
world={'countries':['philippines','usa','germany'],'capitals':['manila','washington dc','berlin'],
       'pop':[100,300,80]}
world_df=pd.DataFrame(world)
display(world_df)
# * Changing labels
world_df.index = ['asia','north america','europe']
display(world_df)

# & type function used to identify the type of objects
print(type(world_df))
# & Sub dataframe
world_df[['countries','pop']]

# & loc - label based
# & iloc - integer-position based
display(world_df.loc[['asia']])
display(world_df.iloc[[1]])

display(world_df.loc[:,['capitals','pop']])
display(world_df.iloc[[1,2],[0,2]])
#Storing the columns in brackets will convert the series into a drafrrame
```

# Logic, Control Flow and Filtering

### Comparison and Boolean Operators

```python
import numpy as np
bmi=30
# * NumPy has its own logical operators
print(np.logical_and(bmi>21 , bmi<40))
print(np.logical_or(bmi>21, bmi<25))
print(np.logical_not(bmi==25))

# * Filtering pandas DataFrame
overpopulated=world_df['pop']>80
display(overpopulated)
display(world_df[overpopulated])

#or if you want as a column
world_df['overpopulated']=overpopulated
display(world_df)
```

# Loops

```python

# * for loop
# & enumerate() - returns the index and value of an iterable --> converts the iterable to an enumerated object
fam=[1.73,1.45,1.65,1.45]
for index,height in enumerate(fam):
    print("index ",index,': ',str(height))
```

```python

# * Looping through the different data structures
# & dictionary
world={'countries':['philippines','usa','germany'],'capitals':['manila','washington dc','berlin'],
       'pop':[100,300,80]}
for key,value in world.items():
    for a in value:
        print(key,a)

# * Using np.nditer() - Each element of an array is visited using Python’s standard Iterator interface.
meas=np.array([data.gdp_cap,data.life_exp])
for val in np.nditer(meas):
    print(val)

# & iterrows - Iterate over DataFrame rows as (index, Series) pairs. 
for x,row in world_df.iterrows():
   display(x) # Returns the index of the values
   display(row) # Returns the column:value for each entry
for lab,row in world_df.iterrows():
    world_df.loc[lab,'COUNTRY']=row['countries'].upper()
# * modification to only print the country
for x,row in world_df.iterrows():
    display(x)
    display("countries: {}".format(row['countries']))

# & pd.apply - Applying a function on a particular column in an element-wise fashion
world_df['name_length']=world_df['countries'].apply(len)
display(world_df)
```

# Case Study: Hacker Statistics

```python
np.random.seed(11)
coin=np.random.randint(0,10)
print(coin)

# * Random walk - a process for determining the probable location of a point subject to random motions, given the probabilities (the same at each step) of moving some distance in some direction
# Example: Gambler's financial status
cards=[]
outcomes=[]
for x in range(10):
    cards.append(x)
    card=np.random.randint(1,14)
    if card == 9:
        outcomes.append('win')
    if card > 9 :
        outcomes.append('high lose')
    else:
        outcomes.append('low lose')
print(cards)
print(outcomes)

# & Distribution 
```
