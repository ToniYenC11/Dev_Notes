## Hosting a Static Website on Amazon S3

- S3 stores HTMl, CSS, JS pages of static website. Limited to HTTP (Use **Amazon Cloudfront** for HTTPS).

- S3 automatically assigns an endpoint UL that you can use to access the website.

- Benefits:
    - No managed infrastructure
    
    - Scales to handle increasing traffic
    - Provides low-cost option for hosting a static website



1. Create S3 bucket

2. Enable website hosting and grant public read permissions to the content of the bucket, configure the S3 bucket.

3. Upload website content

4. You can now access the website at the endpoint URL that Amazon S3 assigns to it. 
    - The separator character before \<Region\> is either a period (.) or a dash (-).
    
    - The choice is because of geographical (period is used for other regions).

## Computing on AWS

### EC2 Virtualization



- Virtual machines run on top of **hypervisor**, the operating layer that provides EC2 instances with access to tha actual hardware for each instance.

- Some instances use **instance store**, a storage that is physically attached to the host computer and provides temporary block-level storage for use with an instance.
    - if the instance reboots, the data in the instance store persists.
    - If the instance stops, the data is lost and cannot be recovered.

#### Network Conectivity



- **Amazon EBS** provides persistent block storage volumes, which means that the data will be persisted. 

- Amazon EBS optimized instances minimize I/O contention between Amazon EBS and other traffic from your instance, resulting in better performance.

### Instance Profile

- Attach to AWS IAM role to an EC2 instnace

- Supplies temporary permissions that applications runnong on the instance use to authenticate when they make calls to AWS resources

- Benefits:
    - You don’t have to store credentials (access key and secret key) locally on the instance, which is a security risk.

    - Credentials are temporary and rotated automatically.
    - You can use a role for multiple instances (i.e. Auto scaling group)

### User Data

- Pass to an instance to perform customization and configuration tasks

- Format depends on OS
    - Shell script for Linux

    - Batch/PowerShell script for Windo- By default, a user data script runs only the first time you launch an instance
