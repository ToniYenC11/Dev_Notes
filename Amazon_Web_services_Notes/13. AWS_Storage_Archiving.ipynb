# Cloud Storage Overview

- **Cloud Storage** stores data on the internet through a provider that manages and operates data storage as a service.

- Cloud storage formats
    - **object** - Each piece of data is an object with a UID.

    - **file** - Data is organized hierarchically with files and folders.
    - **block** - Each piece of data is broken down into blocks with UID.
