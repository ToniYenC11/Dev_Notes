# Amazon Redshift an Data Warehouse

## Data Warehouse

- Central repository of information that can be analyzed to make informed decisions

- Consolidates data from many sources
- Analyze historical data
- Confirm data quality, consistency, and accuracy
- Separate naakytics processing from transactional databases, imporoving the performance of both systems.

### Data Warehouse Architecture

- **Top tier** - Frontend client that presents results through reporting, analysis, and data mining tools.

- **Middle tier** - Consists of the analytics engine that is used to access and analyze the data.

- **Bottom tier** - Database server, where data is stored and loaded.

### Data Warehouse Use Case



## Amazon Redshift

- Fully managed **Data Warehouse** service.

- Run complex analytic queries against *petabytes* of structured data

- Uses sophisticated query optimization, columnar storage on high-performance local disks, and parallel query execution.

## Checkpoint Questions

1. What is Aurora?

    - A fully managed relational database compatible with MySQL and PostgreSQL.

2. What are the two types of instances in Aurora DB cluster?

    - A primary instance and Aurora replicas

3. Why is Aurora ideal for a mobile gaming case
    - Aurora is designed with built-in high availability and fault tolerance, which ensures that mobile gaming applications remain accessible even during traffic spikes or server failures. This is critical for gaming companies, where downtime can lead to a poor user experience and loss of revenue.

    - Aurora can scale to handle large volumes of user data, which is crucial for mobile gaming companies that need to support millions of active users. Its ability to automatically scale with demand ensures that gaming data (e.g., player profiles, in-game progress, leaderboards) can be efficiently stored and accessed.

    - Aurora offers high-performance read and write capabilities with low-latency access to the data, which is important in gaming environments where real-time data updates (such as player actions or game states) are needed.
