## EC2 Auto Scaling

- Ensure application availability by automatically launching/terminating EC2 instances based on scaling options defined.

    - Manual Scaling
    
    - Scheduled Scaling
    - Dynamic scaling
    - Predcitive scaling

### Auto Scaling Concepts

- **Capacity** - Minimum/maximum group size

- **Scaling in and out** - Reduction/increase of CPU use.

- **Instance health** - When an instance is unhealthy, it is terminated to launch a new instance.

- **Termination policy** - Which is terminated during scale-in.

- **Launch template** - ID of AMI, instance type, key pair, security groups, other EC2 instance parameters.

### Auto Scaling Policy

Can be defined through:

- **Amazon CloudWatch** alarms based on alarm level.

- A **target tracking policy** where a metric is evaluated against a target

- Define a **scheduled action**, which sets a new desired capacity value at a specific time. For anticipated traffic spikes periodically.

### [Scheduled Scaling](https://docs.aws.amazon.com/autoscaling/ec2/userguide/ec2-auto-scaling-scheduled-scaling.html)



Steps:

1. Create a scheduled action

2. Specify start time, and the new minimum, maximum, and desired sizes for scaling action

3. Amazon EC2 Auto Scaling updates the group to match the capacity values specified in the scaling action.

### [Dynamic Scaling](https://docs.aws.amazon.com/autoscaling/ec2/userguide/as-scale-based-on-demand.html)

Scales the capacity of your auto scaling group as traffic changes occur.

#### Target Tracking Scaling



- Increase/decrease based on target value for a metric

#### Step scaling

- Increase/decrease based on set of scaling adjustments, aka **step adjustments**. Vary based on size of the alarm breach.

- Does not support *cooldown periods*

#### Simple Scaling

- I/D based on a single scaling adjustment.

- Supports cooldown period between scaling

### Predictive Scaling



- Increase the capacity of your auto scaling group in advance of daily and weekly patterns in traffic flows.

- Uses machine learning models to predict expected traffic. Model re-evaluated evry 24 hours to create a 48-hour forecast.

- Can be used with **Dynamic Scaling**. P forecasts, D scales using target tracking.

See also:

- [Predictive Scaling Settings](https://docs.aws.amazon.com/autoscaling/plans/userguide/gs-specify-custom-settings.html#gs-customize-predictive-scaling)

- [Monitoring and Evaluating Forecasts](https://docs.aws.amazon.com/autoscaling/plans/userguide/gs-create-scaling-plan.html#gs-monitoring-forecasts)

## Auto Scaling Best Practices

- Use a 1-minute frequency in CLoudWatch metric data collection
    - Help ensure a faster response to load changes

    - Scaling on metrics with a 5-minute frequency can result in a slower response time and scaling on stale metric data

    - See https://docs.aws.amazon.com/autoscaling/ec2/userguide/enable-as-instance-metrics.html

- Turn on auto scaling group metrics
    - See https://docs.aws.amazon.com/autoscaling/ec2/userguide/ec2-auto-scaling-cloudwatch-monitoring.html

- Avoid [burstable performance](https://docs.aws.amazon.com/AWSEC2/latest/UserGuide/burstable-credits-baseline-concepts.html) instance types.
    - T3 and T2 instances provide a baseline level of CPU performance, with the ability to burst to a higher level

    - You could exceed the baseline and run out of CPU credits, limiting performance.

See also [Best Practices for Scaling Plans](https://docs.aws.amazon.com/autoscaling/plans/userguide/best-practices-for-scaling-plans.html)
